```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
```

Now an example graph of $x^2$

```python

x=np.linspace(-10,10,100)
plt.plot(x, x**2)
plt.title("Square ")
```

Some examples from notes: square root inverse, $y=1/\sqrt(x+1)$

Domain for this function is $(-1,\infty ] $

To avoid overflow in drawing the graph, we specify `endpoint=False` and write the interval as `5, -1` so that the endpoint is applied to `-1`.

```python
x=np.linspace(5,-1, endpoint=False)
y=1/(np.sqrt(x +1) )
plt.plot(x, y)
plt.title("Function example")
```

Sketch a function 

```python
x=np.linspace(-2,3,6)
y=1+x**2
x1=np.linspace(-2,3)
y1=1+x1**2
plt.plot(x1,y1)
plt.plot(x,y,'o')
plt.title("Sketched Graph $y=1+x^2$")
```

Sketch a directed graph:

$ N=(1,2,3,4,5) $

$ E=\{(1,2) ,(1,3),(3,3),(4,2),(5,4),(4,3)\} $
